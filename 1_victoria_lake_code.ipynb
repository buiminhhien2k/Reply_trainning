```python
filename=open('1_victoria_lake.txt','r')
read_filename=filename.read()
victoria_lake=list(read_filename.split('\n'))
first_line=victoria_lake[0]
list_1st_line=first_line.split()
```

```python
print(list_1st_line)
```

```python
map=victoria_lake[int(list_1st_line[2])+1:-1]
```

```python
mountain_=[]
water=[]
traffic_jam=[]
dirt=[]
rail_cros=[]
standard=[]
highway=[]
railway=[]
import numpy as np
for i in range(len(map)):
    i=len(map)-1-i
    line=map[i]
    for k in range(len(line)):
        character=line[k]
        if character=='#':
            mountain_.append([i,k])
        elif character=='~':
            water.append([i,k])
        elif character=='*':
            traffic_jam.append([i,k])
        elif character=='+':
            dirt.append([i,k])
        elif character=='X':
            rail_cros.append([i,k])
        elif character=='_':
            standard.append([i,k])
        elif character=='H':
            highway.append([i,k])
        elif character=='T':
            railway.append([i,k])
```

```python
mountain_=np.asarray(mountain_)
water=np.asarray(water)
traffic_jam=np.asarray(traffic_jam)
dirt=np.asarray(dirt)
rail_cros=np.asarray(rail_cros)
standard=np.asarray(standard)
highway=np.asarray(highway)
railway=np.asarray(railway)
map_vector=[mountain_,water,traffic_jam,dirt,rail_cros,standard,highway,railway]
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
colorlist = iter(cm.rainbow(np.linspace(0, 1, len(map_vector))))
fig=plt.figure(figsize=(16.3,17.15))
ax = fig.add_subplot()
for vector in map_vector:
    plt.scatter(vector[:,1],vector[:,0],color=next(colorlist),marker='s',s=230)
victoria_lake[1:21]
target_point=[]
for k in victoria_lake[1:21]:
    X,Y,point_=k.split()
    target_point.append([int(X),int(Y)])
for i in target_point:
    plt.scatter(i[0],i[1],marker='X',color='black',s=100)
    ax.text(i[0],i[1],'('+str(i[0])+';'+str(i[1])+')',size=11,rotation=-45,
            color='black',
            verticalalignment='bottom',
            horizontalalignment='right')
start_point=[[7,2],[10,16],[33,8],[20,25],[28,21],[20,16],[38,31],[49,31]]
path={0:['L','LL','LLD','RRRRR'],
       1:['LL','DDD','UUUURR'],
       2:['LL','UUUULLL','RRRRRRRR','DDDDDDLL'],
       3:['R'],
       4:['L','RR','RDDD'],
       5:['U'],
       6:['RR','UUUL'],
       7:['D','LLU']}
answer_1=open('answer_1.txt','w')
for k in range(8):
    a=start_point[k]
    for i in path[k]:
        form='{} {} {}\n'
        answer_1.write(form.format(a[0],a[1],i))
start_point=np.asarray(start_point)
plt.scatter(start_point[:,0],start_point[:,1],marker='X',color='white',s=100,label='True Position')

plt.gca().invert_yaxis()
plt.show()
```
